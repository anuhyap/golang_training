(2)  Visualize the count for each value in the 'Study Hours' column. Note: Study Hours values comes from this >= 6 hours Good >= 4 hours Average < 2 hours Poor.

(3) Using factorplot to show the relationship between numerical and categorical value. Please use figure to show relation between [Study Hours] and [End Semester Percentage]. Percentage is calculated based on the following:

if percentage >=80 then Best
If percentage >= 60 but less than 80 then Very Good
If percentage >= 45 but less than 60 then Good
If Percentage >= 30 but less than 45 then Pass
If Percentage < 30 then Fail

(4) Get the correlation from the data, then plug it to heatmap function and show it as following:



```python
#upload the  file downloaded 
from google.colab import files
uploaded = files.upload()
```

```python
from pandas.core.algorithms import unique
#code

import pandas as pd
import io
import numpy as np
import matplotlib.pyplot as plot
import seaborn as sea

#1
df = pd.read_csv(io.StringIO(uploaded['lab-assignment-04-student-performance.csv'].decode('utf-8')))
print("Description of all the columns: for each colmun, count number of rows, number of unqiue element, the frequency of each unqiue element.")
print(df.describe())

#2
df=df.dropna()
df['sh'].value_counts().plot(kind='barh')
plot.title("STUDY HOURS VISUALIZATION")
plot.xlabel("No of Students")
plot.ylabel("Study Hours")

#3
sea.catplot(x = "sh", y = "esp", data = df,kind="strip")
plot.xlabel("Study Hours")
plot.ylabel("End Semester Percentage")
plot.title("Study Hours RELATIONSHIP WITH End Semester Percentage")
plot.show()


#4
#print("Correlation Heatmap")
for cols in df.columns:
    df[cols] = df[cols].astype('category')
    catogery_cols = df.select_dtypes(['category']).columns
    df[catogery_cols] = df[catogery_cols].apply(lambda x: x.cat.codes)
corr=df.corr()
sea.heatmap(corr,annot=True)
plot.title("Correlation Heatmap")

```
